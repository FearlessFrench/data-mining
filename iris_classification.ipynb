Read data file

```python
# Load libraries
import pandas as pd
cols = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Class']
dataset = pd.read_csv("iris.csv", names=cols)
dataset.head()
```

Specify Columns as Features and the last Column as Class

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 4].values
print(X)
print(y)
```

Split Training dataset and Test dataset

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, stratify=y)
```

Train Model

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_train,y_train)
```

Predict Test Data using Learned Model

```python
y_pred = model.predict(X_test)
print(y_pred)
```

Evaluate predicted class label with actual class label using Accuracy.

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

Evaluate the performance of model using Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

Evaluate test results with Precision, Recall, F1-Score

```python
from sklearn.metrics import classification_report
print (classification_report(y_test, y_pred))
```

Import the Libaries of Classifiers

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report
```

Build the Models and Test the Models using Other Classifier Algorithms

```python
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
models = []
models.append(('KNN-5', KNeighborsClassifier()))
models.append(('KNN-10', KNeighborsClassifier(n_neighbors=10)))
models.append(('Tree', DecisionTreeClassifier()))
models.append(('GaussianNB', GaussianNB()))  # GaussianNB

# evaluate each model in turn
results = []
names = []
scoring = 'accuracy'
for name, clf in models:
  model = clf
  model.fit(X_train,y_train)
  Predited_class = model.predict(X_test)
  accuracy=accuracy_score(y_test, Predited_class)
  results.append(accuracy)
  names.append(name)
  msg = "%s: %f " % (name, accuracy)
  print(msg)
# boxplot algorithm comparison
fig = plt.figure()
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.bar(names,results)

plt.show()
```
